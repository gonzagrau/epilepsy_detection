# Modelos

Todos los modelos cuentan principalmente siempre de 3 "comandos" que vamos a utilizar:
1. El primer comando es llamar al clasificador con sus **hiperparámetros** (parámetros ajustables de cada modelo que permiten controlar el proceso de entrenamiento del mismo; por ejemplo: profundidad del árbol de decisión o cantidad de vecinos para kNN)
2. **Entrenamiento** del modelo con los datos y sus outputs (mediante el atributo `fit`)
3. Pedir al modelo (una vez entrenado) que **prediga** el output de una muestra (mediante el atributo `predict`)

## Árboles de Decisión

image.png

[DecisionTreeClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html?highlight=decision%20tree#sklearn.tree.DecisionTreeClassifier)

El primer tipo de técnica de clasificación que vamos a utilizar es un "Árbol de Decisión". Dado un conjunto de datos se fabrican diagramas de construcciones lógicas, muy similares a los sistemas de predicción basados en reglas, que sirven para representar y categorizar una serie de condiciones que ocurren de forma sucesiva, para la resolución de un problema.

El árbol se puede crear con distintas formas y estructuras, para empezar a ver su funcionamiento vamos a crearlo con los hiperparámetros default que vienen por librería.

```python
from sklearn.tree import DecisionTreeClassifier
clf_DT = DecisionTreeClassifier(random_state=0) #llamo al clasificador con sus hiperparámetros default
clf_DT.fit(X,y) #entreno al modelo con todos los datos de entrada "X" y sus respectivos outputs "y"
predicted = clf_DT.predict([x1, x2, x3]) # predigo lo que me arroja el modelo entrenado
```

Como verán, el único hiperparámetro que "aclaré" del `DecisionTreeClassifier` es el `random_state`, el cual está asociado a los algoritmos pseudoaleatorios que están asociados a la formación de este árbol. En otras palabras: aclaro este "estado de aleatoriedad" para que la formación del árbol sea aleatoria pero siempre igual de aleatoria cuando le ponga "run" a esta línea.

Ahora comparemos lo que predijo el árbol con las etiquetas reales de estas 3 muestras:

```python
print('REALES', [y1,y2,y3] )
print('PREDICHAS', predicted)
```

Como verán, el arbol de decisión predijo correctamente los output de estas 3 muestras.

A continuación se puede visualizar el árbol decisión que se creó con la función `plot_tree` del módulo `tree` de Scikit-Learn:

```python
from sklearn import tree
__ = tree.plot_tree(clf_DT)
```

El árbol que creamos recién, tiene una morfología y estructura que depende de los hiperparámetros defaults que vienen por librería que son los siguientes (aclarados en la documentación de `DecisionTreeClassifier`):

image.png

El detalle de qué significa cada parámetro escapa a la materia, pero para poder "entender" un poco cómo manejar estos hiperparámetros, cambiemos la estructura del árbol y creemos otro árbol cambiando al hiperparámetro `max_depth`:

```python
otro_clf_DT = DecisionTreeClassifier(max_depth=3,random_state=0) # Acá deje el mismo random_state pero le aclaré un max_depth de 3
otro_clf_DT.fit(X,y)
otro_predicted = otro_clf_DT.predict([x1, x2, x3])
print('REALES', [y1,y2,y3] )
print('PREDICHAS', otro_predicted)
```

Como verán, cree un árbol "menos profundo" que el original (vean el `plot_tree` de acá abajo que muestra como la profundidad de este árbol es de 3 y el del primero es más profundo). El primer arbol creado tiene el parámetro por default `None` que significa que los nodos se expanden hasta que todas las hojas sean puras o hasta que todas las hojas contengan menos de `min_samples_split` (esto lo dice la documentación de DecisionTreeClassifier). Entonces, al crear un árbol menos profundo puede que se haya creado un árbol "menos" preciso en sus predicciones (porque no toma en cuenta todas las decisiones lógicas necesarias) y por eso en la muestra 2 se equivoca con su predicción.

```python
__ = tree.plot_tree(otro_clf_DT)
```

La profundidad del árbol en este caso ideal no es ni la del `clf_DT` ni la de `otro_clf_DT`. La profundidad ideal debería ser el equilibrio entren un árbol que prediga lo más correctamente con la menor profundidad posible. Hacer esta optimización de morfología del árbol, escapa a la materia, pero es algo que los que los que trabajan en Machine Learning siempre estudian y analizan.

## k -Nearest Neighbour (kNN)

image.png

[KNeighborsClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html#sklearn.neighbors.KNeighborsClassifier)

La técnica de kNN (o k nearest neighbours o k vecinos más cercanos) estima el valor de la probabilidad de que un elemento pertenezca a cierta clase según su ubicación espacial dada por las características de los datos. La decisión de que un dato pertenezca a una clase, depende de su cercanía a los $k$ vecinos que tenga más cercanos.

Por eso comencemos creando un clasificador basado en kNN que tome en cuenta los 3 vecinos más cercanos de cada dato que ingrese, lo entrenamos y vemos qué predice para las 3 muestras que sacamos del dataset.


```python
from sklearn.neighbors import KNeighborsClassifier

clf_kNN = KNeighborsClassifier(n_neighbors=3) #aclaro que quiero 3 vecinos únicamente para hacer la clasificación
clf_kNN.fit(X, y)
predicted = list(clf_kNN.predict([x1,x2,x3]))
print('REALES', [y1,y2,y3] )
print('PREDICHAS', predicted)
```

Como verán, el clasificador basado en kNN predijo correctamente los output de estas 3 muestras.

A continuación les voy a mostrar un clasificador basado en kNN que es más cercano a lo que se trabaja en la realidad y lo que hace es tomar en cuenta varios vecinos cercanos pero también ajustando el hiperparámetro `weights` haciendo que la ubicación espacial de los puntos tengan más peso según su distancia a la nueva muestra.

```python
otro_clf_kNN = KNeighborsClassifier(n_neighbors=10, weights='distance') #aumente la cantidad de vecinos y cambié el parámetro de weights
otro_clf_kNN.fit(X, y)
otro_predicted = list(otro_clf_kNN.predict([x1,x2,x3]))
print('REALES', [y1,y2,y3] )
print('PREDICHAS', otro_predicted)
```

Predice las tres muestras también correctamente.

## Support Vector Machine

image.png

[SVC](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html)

SVM = Máquina de Vectores de Soporte/ Support Vector Machine.

Este tipo de clasificador lo que hace es dado un conjunto de muestras y sus etiquetas, entrena una SVM (máquina de vectores de soporte) para construir un modelo que prediga la clase de una nueva muestra. Una SVM es un modelo que representa a los puntos de muestra en el espacio, y puede separar las clases a 2 espacios lo más amplios posibles mediante un hiperplano de separación definido como el vector entre los 2 puntos, de las 2 clases, más cercanos al que se llama *vector de soporte*. Cuando las nuevas muestras se ponen en correspondencia con dicho modelo, en función de los espacios a los que pertenezcan, pueden ser clasificadas a una o la otra clase. En otras palabras, una SVM construye un hiperplano o conjunto de hiperplanos en un espacio de dimensionalidad muy alta que permite clasificar ya que hace una correcta separación entre las muestras según sus clases.

Creamos una SVM que clasifica con sus parámetros por default, lo entrenamos con el dataset y vemos qué predice de las 3 muestras

```python
from sklearn import svm
clf_SVC = svm.SVC()
clf_SVC.fit(X, y)
predicted = list(clf_SVC.predict([x1,x2,x3]))
print('REALES', [y1,y2,y3] )
print('PREDICHAS', predicted)
```

Como verán, el clasificador basado en SVM predijo correctamente 2 de las 3 muestras.

SVM es uno de los algoritmos más potentes, por ende suele ser el que mejores resultados arroja. En este caso estamos en un ejemplo que no sucede así pero esto es por su lógica de funcionamiento. Caimos justo en una muestra que "cuesta" predecirla con esta lógica que utiliza SVM, pero en general funciona mucho mejor que kNN y los árboles de decisión.

Los hiperparámetros internos de SVM también hará que mejore su funcionamiento, pero como el entendimiento de qué significa cada hiperparámetro escapa a la materia, recomendamos leer bibliografía para poder "mejorar" al modelo si así se lo desee.

# Evaluación de Modelos

Hasta ahora, lo que hicimos fue entrenar modelos con todo el set de datos y agarrar 3 muestras que se me ocurrieron para ver cómo predecían mis clasificadores a estas muestras.

Como eran muestras que el clasificador ya vió, era altamente probable que al volver a ingresar al modelo para predecirse, esta predicción sea la correcta. Por eso, sería más correcto probar al modelo con muestras que mi clasificador no vio. Por eso (como se vió en la teoría), separemos al dataset en dos sets de datos para trabajar:
1. Set de entrenamiento: Entrena al modelo
2. Set de pruebta/testeo: El modelo no ve a estos datos, pero vemos qué predice para ver "qué tan bien" se entrenó

[train_test_split](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html?highlight=train_test_split#sklearn.model_selection.train_test_split)

```python
from sklearn.model_selection import train_test_split
```

Esta función nos va permitir formar el set de datos de entrenamiento y el set de datos de prueba, de manera pseudo-aleatoria y separar los datos y las etiquetas de manera ordenada.

!!! IMPORTANTE: Vean los parámetros `test_size` y `train_size` de esta función. Pueden aclarar cómo separar al dataset. Por default estos valores hacen que las muestras se dividan un 75% para entrenar al modelo y un 25% para hacer el testeo.

[plot_confusion_matrix](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.plot_confusion_matrix.html?highlight=plot_confusion_matrix)

```python
from sklearn.metrics import plot_confusion_matrix
```

Esta función nos va a permitir armar la matriz de confusión que es la que nos arroja un resumen de cómo está funcionando el modelo y qué tan bien está prediciendo las muestras que no vio al entrenarse.

image.png

```python
class_names = dataset.target_names

# Divido los datos en un conjunto de entrenamiento y un conjunto de prueba
# (vean la documentación ya que los datos son arrojados en el orden que los estoy recibiendo acá)
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)

# Ejecuto los clasificadores y los entreno (son los que creamos antes que tienen sus hiperparámetros "sin tocar")
clf_DT.fit(X_train, y_train)
clf_kNN.fit(X_train, y_train)
clf_SVC.fit(X_train, y_train)
clf_RF.fit(X_train, y_train)
```

Ahora graficamos la matriz de confusión según lo entrenado

```python
# Matriz de confusión
disp = plot_confusion_matrix(clf_DT, X_test, y_test, display_labels=class_names, cmap=plt.cm.Blues)
disp.ax_.set_title("Matriz de confusión para Árbol de Decisión")
print(disp.confusion_matrix)
print()
plt.show()
```

```python
# Matriz de confusión
disp = plot_confusion_matrix(clf_kNN, X_test, y_test, display_labels=class_names, cmap=plt.cm.Blues)
disp.ax_.set_title("Matriz de confusión para kNN")
print(disp.confusion_matrix)
print()
plt.show()
```

```python
# Matriz de confusión
disp = plot_confusion_matrix(clf_SVC, X_test, y_test, display_labels=class_names, cmap=plt.cm.Blues)
disp.ax_.set_title("Matriz de confusión para SVM")
print(disp.confusion_matrix)
print()
plt.show()
```

```python
# Matriz de confusión
disp = plot_confusion_matrix(clf_RF, X_test, y_test, display_labels=class_names, cmap=plt.cm.Blues)
disp.ax_.set_title("Matriz de confusión para Random Forest")
print(disp.confusion_matrix)
print()
plt.show()
```

Como verán, acá sí se puede apreciar que SVM y RF son los que menos muestras del set de testeo clasifica mal (diagonal secundaria de la matriz) y más muestras del set de testeo clasifica bien (diagonal principal de la matriz).

Esta matriz hace un resumen, pero para "medir" mejor este rendimiento de los clasificadores y poder seleccionar qué modelo "funciona" mejor, hay que medir sus métricas.

# Selección de modelos

### **[MÉTRICAS](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.metrics)**

Las principales métricas a medir son las siguientes que se basan en los resultados obtenidos en la matriz de confusión. Se puede medir con la matriz de confusión que creamos (es decir, a partir del set de testeo) pero, en algunos casos, también sirve crearla con la matriz de confusión con los datos de entrenamiento. Recomendamos únicamente medir esto con los datos de testeo a menos que quieran realizar un análisis específico sobre el conjunto de entrenamiento.

image.png

[accuracy_score](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html#sklearn.metrics.accuracy_score)

Esta función nos va a dar el parámetro "accuracy" que es el que cuenta cuántas muestras fueron clasificadas bien sobre el total de muestras.

```python
from sklearn.metrics import accuracy_score

y_pred_DT = clf_DT.predict(X_test)
y_pred_kNN = clf_kNN.predict(X_test)
y_pred_SVC = clf_SVC.predict(X_test)
y_pred_RF = clf_RF.predict(X_test)

acc_DT = accuracy_score(y_test, y_pred_DT)
acc_kNN = accuracy_score(y_test, y_pred_kNN)
acc_SVC = accuracy_score(y_test, y_pred_SVC)
acc_RF = accuracy_score(y_test, y_pred_RF)
```

```python
print('Accuracy score de Decision Tree', acc_DT)
print('Accuracy score de kNN', acc_kNN)
print('Accuracy score de SVM', acc_SVC)
print('Accuracy score de Random Forest', acc_RF)
```

Acá podemos observar que RF y SVM son los que obtienen un resultado más alto.

Para hablar de que un modelo tiene un "buen" accuracy, tenemos que estar hablando de valores mayores al 0.85, pero fijense que en este caso nos sirve para comparar los modelos y afirmar que "Random Forest" es el que hace mayor cantidad de predicciones correctas.

[classification_report](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html#sklearn.metrics.classification_report)

```python
from sklearn.metrics import classification_report
```

Esta función nos va a dar un reporte con muchas métricas del funcionamiento de los clasificadores (precision, recall, f1-score). Nos va a servir para hacer un mejor análisis de qué clasificador es el que mejor funciona.

```python
print('DECISION TREE REPORT')
print(classification_report(y_test, y_pred_DT, target_names=class_names))
print('\n kNN REPORT')
print(classification_report(y_test, y_pred_kNN, target_names=class_names))
print('\n SVM REPORT')
print(classification_report(y_test, y_pred_SVC, target_names=class_names))
print('\n RANDOM FOREST REPORT')
print(classification_report(y_test, y_pred_RF, target_names=class_names))
```

### **[COMPARACIÓN](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.model_selection)**

Hasta ahora, lo que hicimos fue separar al dataset en un 75% para entrenarlo y 25% para testeo, de manera aleatoria y ordenada. Pudo haber pasado, que agarrarmos una importante proporción de ese 75% de los datos sean "malos" para entrenarlo (díficil pero puede suceder). Como para no caer en un caso específico, lo "más correcto" es hacer estas separaciones de manera aleatoria pero varias veces de maneras distintas. Para eso existe una forma de hacerlo que se llama "Cross Validation" que es agarrar varias veces distintas partes aleatorias distintas del set de datos para formar al set de testeo y de entrenamiento.

image.png

```python
from sklearn.model_selection import cross_val_score
```

Esta función nos permite hacer "Cross Validation" y además medir el accuracy en cada división aleatoria.

```python
kNN_score_cv = cross_val_score(clf_kNN, X, y, cv=5)
# Acá hice lo mismo que antes pero 5 veces de maneras distintas aleatorias para el algoritmo de kNN
print(kNN_score_cv)
```

Pueden ver que los 5 resultados son distintos pero el primer caso da un número que claramente es más chico que el real, es decir, separe en testeo y entrenamiento en un caso que da un rendimiento bajo. Las otras 4 particiones que se hicieron, me dan una indicación que en realidad puede tener un mejor rendimiento.

Por eso la mejor forma de decir el accuracy de un clasificador es dar la media de estos resultados con su desvío:

```python
print("Accuracy para kNN: %0.2f (+/- %0.2f)" % (kNN_score_cv.mean(), kNN_score_cv.std() * 2))
```

Por otra parte, [cross_validate](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_validate.html?highlight=cross_validation) es otra función muy parecida, pero un poco más versátil e informativa. Además de devolver los `scores`, devuelve algunas métricas que pueden ser útiles, como por ejemplo: tiempo de entrenamiento, los modelos entrenados, etc. Además, permite evaluar más de una métrica.

```python
from sklearn.model_selection import cross_validate
```

```python
cv_DT = cross_validate(clf_DT, X, y, cv=3)
cv_kNN = cross_validate(clf_kNN, X, y, cv=3)
cv_SVC = cross_validate(clf_SVC, X, y, cv=3)
cv_RF = cross_validate(clf_RF, X, y, cv=3)

print(cv_DT['test_score'])
print(cv_kNN['test_score'])
print(cv_SVC['test_score'])
print(cv_RF['test_score'])
```

Si vamos a cada `cv_clf`. vamos a ver que tienen asociado distintos parámetros que nos pueden indicar cómo se entrenó:

```python
cv_DT
```

```python
cv_RF
```

### Random Search

[RandomizedSearchCV](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.RandomizedSearchCV.html#sklearn.model_selection.RandomizedSearchCV)

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform
```

```python
distributions = {'C': uniform(loc=1, scale=10) }
clf = RandomizedSearchCV(svm.SVC(), distributions, random_state=15)
search = clf.fit(X, y)
```

```python
search.best_params_
```
